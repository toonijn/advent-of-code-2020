```python
neighbors = [(i, j) for i in [-1,0,1] for j in [-1,0,1]]
neighbors.remove((0,0))

def simulate_1(L):
    rows = len(L)
    cols = len(L[0])
    
    R = [['.']*cols for _ in range(rows)]
    changed = False
    
    for i in range(rows):
        for j in range(cols):
            if L[i][j] != '.':
                occupied = 0
                for di, dj in neighbors:
                    if 0 <= i+di < rows and 0 <= j+dj < cols \
                            and L[i+di][j+dj] == '#':
                        occupied += 1
                if occupied >= 4:
                    R[i][j] = 'L'
                elif occupied == 0: 
                    R[i][j] = '#'
                else:
                    R[i][j] = L[i][j]
                changed = changed or R[i][j] != L[i][j]
    return R, changed



def simulate_2(L):
    rows = len(L)
    cols = len(L[0])
    
    R = [['.']*cols for _ in range(rows)]
    changed = False
    
    for i in range(rows):
        for j in range(cols):
            if L[i][j] != '.':
                occupied = 0
                for di, dj in neighbors:
                    si = i+di
                    sj = j+dj
                    while 0 <= si < rows and 0 <= sj < cols and L[si][sj] != "L":
                        if L[si][sj] == '#':
                            occupied += 1
                            break
                        si += di
                        sj += dj
                if occupied >= 5:
                    R[i][j] = 'L'
                elif occupied == 0: 
                    R[i][j] = '#'
                else:
                    R[i][j] = L[i][j]
                changed = changed or R[i][j] != L[i][j]
    return R, changed
```

```python
with open('day11.txt') as d:
    L = list(map(list, map(str.strip, d)))

changed = True
i = 0
while changed:
    i += 1
    L, changed = simulate_2(L)
    
print(f"Na {i} stappen zijn er {sum(r.count('#') for r in L)} stoelen bezet.\n")
print('\n'.join(map(''.join, L)))
```
