```python
with open("day22.txt") as d:
    p1, p2 = d.read().strip().split('\n\n')
    # p1, p2 = """Player 1:\n9\n2\n6\n3\n1\n\nPlayer 2:\n5\n8\n4\n7\n10""".split('\n\n')
    p1 = list(map(int, p1.split('\n')[1:]))
    p2 = list(map(int, p2.split('\n')[1:]))
    
def recursive_combat(p1, p2):
    ### returns p1_wins, p1, p2
    print(f"Playing: {p1} <-> {p2}")
    sets_p1 = set()
    while p1 and p2:
        # print(p1, p2)
        if tuple(p1) in sets_p1:
            print(f"Win by p1")
            return True, p1, p2
        sets_p1.add(tuple(p1))
        
        a = p1.pop(0)
        b = p2.pop(0)
        
        if a > len(p1) or b > len(p2):
            p1_wins = a > b 
        else:
            p1_wins = recursive_combat(p1[:a], p2[:b])[0]
        
        if p1_wins:
            p1 += [a, b]
        else:
            p2 += [b, a]
    
    print(f"Win by {'p1' if p1 else 'p2'}")
    return bool(p1), p1, p2

p1_wins, p1, p2 = recursive_combat(p1, p2)
p1_wins
```
